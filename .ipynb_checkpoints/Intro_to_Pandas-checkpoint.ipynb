## Merge, Join and Concatenate

```python
import pandas as pd
```

```python
df1 = pd.read_csv('dataset1.csv')
df1
```

```python
df2 = pd.read_csv('dataset2.csv')
df2
```

```python
df1.merge(df2)
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, how="outer")
```

```python
pd.merge(df1, df2, on="FellowshipID", how="outer")
```

```python
pd.merge(df1, df2, on="FellowshipID", how="inner")
```

```python
pd.merge(df1, df2, on="FellowshipID", how="left")
```

```python
pd.merge(df1, df2, on="FellowshipID", how="right")
```

```python
pd.merge(df1, df2, on=["FellowshipID", "FirstName"])
```

```python
pd.merge(df1, df2, on=["FellowshipID", "FirstName"], how="outer")
```

```python
pd.merge(df1, df2, on=["FellowshipID", "FirstName"], how="left")
```

```python
pd.merge(df1, df2, on=["FellowshipID", "FirstName"], how="right")
```

```python
pd.merge(df1, df2, how="cross")
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], join="inner")
```

```python
pd.concat([df1, df2], join="outer")
```

```python
pd.concat([df1, df2], join="outer", axis=1)
```
